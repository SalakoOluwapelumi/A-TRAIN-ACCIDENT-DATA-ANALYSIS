```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This is a data that shows a death case of passengers in a train accident, 
# according to their names, sex, fare, Pclass , survivors.. 
death_case = pd.read_csv('train.csv')
death_case
```

```python
# iin this ata set, some data were not properly recorded (due to inablity to get the proper record),
# so they were set to NaN(Not a number), which is not proper to work with in a dataset. So i decided to remove any data having NaN in it

death_case.dropna()
```

```python
# In order not to cause any alteration in the dataset, it is better to fill al NaN to 0, so all NaN in the dataset will be set to 0 
pd.set_option('display.max_rows', None)
death_case.fillna(0)
```

```python
# This is Bar Chart showing the numbers and range of the survivor in the train accident
survivals = death_case.groupby('Sex')['Survived'].sum()
colours = ['blue', 'green']
plt.bar(survivals.index, survivals.values, color = colours)
plt.title('Survivors by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of survivors')
plt.show()

# Over 200 female survived and over an 100 male survived 
```

```python
# Pie chart representation of the survivors sex and number range
colours = ['brown', 'blue']
plt.pie(survivals.values, labels = survivals.index, colors = colours)
plt.title('Survivors by Gender')
plt.show()
```

```python
# This histogram shows the number range of those who were in the train accident with respect to their Age
colours = ['blue']
plt.hist(death_case['Age'], color = colours)
plt.title('Ages of pessengers')
plt.xlabel('Ages')
plt.ylabel('Range')
plt.show()

# This shows that up to 175 people in the train accident were 20 years old, over 150 peoplein the train were 30 years, and so on.
# So this can be concluded that youne people with age range(10-50) were more in the train as compared to the adults of with age range(60-80)
```

```python
# This scatter diagram shows the fare of the people by the class in wich they are.
plt.scatter(death_case['Pclass'], death_case['Fare'])
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()

# This shows that a lot of people were more in the class of 1 than the other classes.
```

```python
# Tis line chart shows the  number of people in the train who survived according to their Age 
survivals = death_case.groupby('Age')['Survived'].sum()
plt.plot( survivals.index, survivals.values, marker ='o', color = 'blue')
plt.title('Survivors by Age')
plt.xlabel('Age')
plt.ylabel('Range')
plt.show()

# This shows that the age of of 20 - 35 survived more than the other people age ranges
```
